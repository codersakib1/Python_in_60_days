```python
import PyPDF2
path = 'ML_Engineer_Roadmap_2024_Abdullah_Al_Mamun.pdf'
with open(path, 'rb') as file:
    reader = PyPDF2.PdfReader(file)

    #first_page = reader.pages[0].extract_text()
    #print(first_page)

    for page in reader.pages:
        text = page.extract_text()
        print(text)
        print('\n')
```
